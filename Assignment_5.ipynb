```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**# 1. Write a Python programming to display a bar chart of the popularity of
programming Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7**

```python
programming = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2,17.6,8.8,8,7.7,6.7]

plt.bar(programming,popularity,color = 'b')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.minorticks_on()
plt.xlim(-0.4)
plt.grid(True,alpha = 0.2,ls = '-',which = 'major',color = 'r',lw = 1)
plt.grid(True,alpha = 0.1,ls = '-',which = 'minor',color = 'b',lw = 1)
plt.show()
```

**# 5. Write a Python programming to display a bar chart of the popularity of
programming Languages. Attach a text label above each bar displaying its
popularity (float value).
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7**

```python
programming = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2,17.6,8.8,8,7.7,6.7]

plt.bar(programming,popularity,color = 'b')
bar_container = plt.bar(programming, popularity)
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.bar_label(bar_container,fmt='{:,.5f}')
plt.minorticks_on()
plt.xlim(-0.4)
plt.grid(True,alpha = 0.2,ls = '-',which = 'major',color = 'r',lw = 1)
plt.grid(True,alpha = 0.1,ls = '-',which = 'minor',color = 'b',lw = 1)
plt.show()
```
