```python
# 3. Write a Python program to draw a line using given axis values taken from a
# text file, with suitable label in the x axis, y axis and a title.
# Test Data:
# test.txt
# 1 2
# 2 4
# 3 1

data = open('test.txt','r')
lines = data.readlines()
xlst=[]
ylst =[]

for line in lines:
    line = line.split()
    x =(int(line[0]))
    y =(int(line[1]))
    xlst.append(x)
    ylst.append(y)


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot with text file!')
plt.xlim(left = min(xlst),right =max(xlst))
plt.ylim(bottom = min(ylst),top = max(ylst))
plt.plot(xlst,ylst,c='r')
plt.show()
```

```python
# 6. Write a Python program to plot two or more lines with legends, different
# widths and colors.

x = [10,20,30]
y = [20,40,10]
a=[10,20,30]
b=[40,10,30]
plt.plot(x,y,'b',linewidth = 2.0)
plt.plot(a,b,'r',linewidth = 4.0)
plt.legend(['line1-width-2','line2-width-4'])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('plot with width')

plt.show()
```

```python
# 7. Write a Python program to plot two or more lines with different styles.

x = [10,20,30]
y = [20,40,10]
a=[10,20,30]
b=[40,10,30]
plt.plot(x,y,'b',linewidth = 2.0,linestyle ='dotted')
plt.plot(a,b,'r',linewidth = 4.0,linestyle ='dashed')
plt.legend(['line1- dotted','line2 - dashed'])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('plot with style')
```

```python
# 8. Write a Python program to plot two or more lines and set the line markers.

x = [1,4,5,6,7]
y = [2,6,3,6,3]
plt.plot(x,y,'r',ls='-.',lw='2',ms='10',marker='o',mfc='b',mec='b')
plt.legend(['line1'])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('plot with markers')
plt.show()
```

```python

# 14. Write a Python program to display the grid and draw line charts of the
# closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016.
# Customized the grid lines with rendering with a larger grid (major grid) and a
# smaller grid (minor grid).Turn on the grid but turn off ticks.

df = pd.read_csv('fdata.csv')
plt.plot('Date','High','r',ls='-.',data=df,marker='o',markerfacecolor='blue',ms=8,)
plt.grid(ls = '-.',c='b',alpha = 0.2,which = 'minor')
plt.grid(ls = '-',c='r',alpha = 0.7,which = 'major')
plt.minorticks_on()
plt.show()
```
