# **14. Write a Pandas program to create a plot of adjusted closing prices, thirty days and forty days simple moving average of Alphabet Inc. between two specific dates.**

```python
sd = pd.to_datetime('2020-04-01')
ed = pd.to_datetime('2020-09-30')  #set startdate and enddate
df['Date'] = pd.to_datetime(df['Date'])   #set the date column in df in datetime format from the regular string format
newdf = (df['Date']>=sd) & (df['Date'] <=ed)  #taking dates only between sd to ed
df1 = df.loc[newdf]   # newdf is mask and df2 is filtered df
df2 = df1.set_index('Date')

close_px = df2['Adj Close']
df2['SMA_30_days'] = df2.iloc[:,4].rolling(window = 30).mean()
df2['SMA_40_days'] = df2.iloc[:,4].rolling(window = 40).mean()

plt.figure(figsize =[6,4])
plt.grid(True)
plt.title('Historical stock prices of Alphabet Inc. [01-04-2020 to 30-09-2020]\n',fontsize=14, color='black')
plt.xlabel('Date',fontsize=12, color='black')
plt.ylabel('Adj Close', fontsize=12, color='black')
plt.plot(df2['Adj Close'],label = 'Adj Close')
plt.plot(df2['SMA_30_days'],label = 'SMA_30_days')
plt.plot(df2['SMA_40_days'],label = 'SMA_40_days')
plt.legend(loc = 2)
plt.show()
```

# **15. Write a Pandas program to create a plot of adjusted closing prices, 30 days simple moving average and exponential moving average of Alphabet Inc. between two specific dates.**

```python
sd = pd.to_datetime('2020-04-01')
ed = pd.to_datetime('2020-09-30')  #set startdate and enddate
df['Date'] = pd.to_datetime(df['Date'])   #set the date column in df in datetime format from the regular string format
newdf = (df['Date']>=sd) & (df['Date'] <=ed)  #taking dates only between sd to ed
df1 = df.loc[newdf]   # newdf is mask and df2 is filtered df
df2 = df1.set_index('Date')

close_px = df2['Adj Close']
df2['SMA_30_days'] = df2.iloc[:,4].rolling(window = 30).mean()
df2['EMA_40_days'] = df2.iloc[:,4].ewm(span = 20, adjust = False).mean()

#ewm(): This is a pandas method used to compute the Exponentially Weighted Moving (EWM) average

plt.figure(figsize =[6,4])
plt.grid(True)
plt.title('Historical stock prices of Alphabet Inc. [01-04-2020 to 30-09-2020]\n',fontsize=14, color='black')
plt.xlabel('Date',fontsize=12, color='black')
plt.ylabel('Adj Close', fontsize=12, color='black')
plt.plot(df2['Adj Close'],label = 'Adj Close')
plt.plot(df2['SMA_30_days'],label = 'SMA_30_days')
plt.plot(df2['EMA_40_days'],label = 'EMA_40_days')
plt.legend(loc = 2)
plt.show()
```

# **17. Write a Pandas program to create a plot to visualize daily percentage returns of Alphabet Inc. stock price between two specific dates.**

```python
sd = pd.to_datetime('2020-04-01')
ed = pd.to_datetime('2020-05-30')
df['Date'] = pd.to_datetime(df['Date'])
newdf = (df['Date']>=sd) & (df['Date'] <=ed)
df1 = df.loc[newdf]
df2 = df1.set_index('Date')
df2['Close'] = df2['Close'].pct_change()
df2.dropna()
print(df2['Close'].head())

plt.figure(figsize = (10,5))

plt.suptitle(' daily percentage returns of Alphabet Inc. stock price', fontsize=12, color='black')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("$ price", fontsize=12, color='black')
df2['Close'].plot(color='red',marker='o',mfc = 'black',mec = 'black');
plt.grid()
plt.show()
```

# **18. Write a Pandas program to plot the volatility over a period of time of Alphabet Inc. stock price between two specific dates.**

```python
sd = pd.to_datetime('2020-04-01')
ed = pd.to_datetime('2020-05-30')
df['Date'] = pd.to_datetime(df['Date'])
newdf = (df['Date']>=sd) & (df['Date'] <=ed)
df1 = df.loc[newdf]
df2 = df1.set_index('Date')
df2['Close'] = df2['Close'].rolling(window = 2).std()
print(df2['Close'].head())

plt.figure(figsize = (10,5))

plt.suptitle('volatility over a period of time of Alphabet Inc. stock price', fontsize=12, color='black')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("$ price", fontsize=12, color='black')
df2['Close'].plot(color='red',marker='o',mfc = 'black',mec = 'black');
plt.grid()
plt.show()
```

# **19. Write a Pandas program to create a histogram to visualize daily return distribution of Alphabet Inc. stock price between two specific dates.**

```python
df = pd.read_csv("alphabet_stock_data.csv")
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-9-30')
df['Date'] = pd.to_datetime(df['Date'])
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df1 = df.loc[new_df]
df2 = df1[['Date', 'Adj Close']]
df3 = df2.set_index('Date')
daily_changes = df3.pct_change(periods=1)
sns.distplot(daily_changes['Adj Close'].dropna(),bins=100,color='purple')
plt.suptitle('Daily % return of Alphabet Inc. stock price,\n01-04-2020 to 30-09-2020', fontsize=12, color='black')
plt.grid(True)
plt.show()
```
